<font color='#4682B4'>

***(1)***

**RÂ² = 1 - (SSR / SST)**

Where:   
SSR: Sum of Squares Residuals (also called Sum of Squared Errors or SSE), unexplained variance in the dependent variable by the regression model.       
SST: Total Sum of Squares, which measures the total variance in the dependent variable.


**Adj RÂ² = 1 - [(1-RÂ²).(n-1)]/(n-p)**

Where:  
RÂ² is the regular R-squared | n is the number of observations | p is the number of features in the model

The adjusted R-squared is a modified version of R-squared that has been adjusted for the number of predictors in the model. The adjusted R-squared increases only if the new term improves the model more than would be expected by chance. It decreases when a predictor improves the model by less than expected by chance. The adjusted R-squared can be negative, but itâ€™s usually not.  It is always lower than the R-squared.  
Also, by not including unimportant features, we are able to keep the model simple and more interpretable. These advantages makes Adj RÂ² more desirable to use.
    
</font>

<font color='#4682B4'>

***(2)***
    

**Yes, Overfitting can happen in linear regression when -**

**1) Highy complex model** - if we use a polynomial regression with a high degree (e.g., quadratic or cubic) when the true relationship between the variables is linear. This introduces complexity to the model. The model will try to capture every fluctuation and noise in the data, resulting in a highly flexible curve that passes through every data point. While this may yield a perfect fit on the training data, it often fails to generalize well to new, unseen data.       
**2) Feature ~ Observations -** Running linear models on a dataset with very few obsevations and features almost equal to observations will lead to overfitting. Imagine a case where you have 2 observations and just one feature. Then by linear regression you can fit a straight line that passes through both the points and there is no error. This linear model most definetely will not perform well on new data.   
**3) Soo much noise! -** A linear model is also prone to overfitting where is a lot of random noise on the training data as it will try to fit the noise and doesn't perform well on the test data

By implementing regularization techniques, such as Ridge Regression or Lasso Regression we can tackle the problem of overfitting. Regularization adds a penalty term to the linear regression cost function, discouraging the model from assigning excessive importance to individual features. Ridge Regression adds a penalty based on the L2-norm of the coefficients, while Lasso Regression uses the L1-norm. These methods can help shrink coefficients toward zero and reduce overfitting.

## Part-2: *(5 pts)*

Use `sklearn.linear_model.Lasso` and `sklearn.linear_model.Ridge` classes to do a 5-fold cross validation using sklearn's `KFold`. For the sweep of the regularization parameter, we will look at a grid of values ranging from  $\alpha=10^{-6}$  to  $\alpha=10^{6}$.In Python, you can consider this range of values as follows: `alpha = 10**numpy.linspace(-6, 6, 100)` so that you can generate 100 uniform values between -6 to 6 as power series.

Fit the 2 regression models (Lasso and Ridge) with scaled data and report the best chosen $\alpha$ based on cross validation as well as the corresponding scoring metric. The cross validation should happen on your training data using MSE as the scoring metric.

```python
# Define number of folds
##  START CODE  ## (1 line of code)
n_folds = 5
##  END CODE  ##
```

```python
# Create KFold from sklearn
##  START CODE  ## (1 line of code)
k_fold = KFold(n_splits=n_folds, shuffle=True, random_state=50)
##  END CODE    ##
```

```python
#Define the alphas as defined in the question
##  START CODE  ## (1 line of code)
alphas = 10**np.linspace(-6, 6, 100)
##  END CODE    ##

lasso_avg_mse = {}
ridge_avg_mse = {}
```

```python
from sklearn.linear_model import Lasso
#For each value of alpha and each fold compute the mean square error
for alpha in alphas:
    avg_mse = 0
    for k, (train, test) in enumerate(k_fold.split(Xscaled_train, y_train)):
        lasso = Lasso(alpha=alpha)
        #Fit the scaled training data to the lasso model
        ## START CODE ## (1 line of code)
        lasso.fit(Xscaled_train[train], y_train[train])
        ## END CODE ##

        #Calculate the average mean sqaured error
        ##  START CODE  ## (1 line of code)
        avg_mse  += mean_squared_error(y_train[test], lasso.predict(Xscaled_train[test]))
        ##  END CODE    ##

    # Take the average mean squared error as metric
    lasso_avg_mse[alpha] = avg_mse / n_folds
```

```python
# Find the best value for alpha with minimum mean squared error
##  START CODE  ## (1 line of code)
best_alpha_lasso = min(lasso_avg_mse, key=lasso_avg_mse.get)
##  END CODE    ##

print(f"Best Alpha: {best_alpha_lasso}, Corresponding Average MSE: {lasso_avg_mse[best_alpha_lasso]}")
```

```python
from sklearn.linear_model import Ridge
#For each value of alpha and each fold compute the mean square error
for alpha in alphas:

    #Instantiate a ridge model with the current alpha
    ##  START CODE  ## (1 line of code)
    ridge = Ridge(alpha=alpha)
    ##  END CODE    ##

    avg_mse = 0

    for k, (train, test) in enumerate(k_fold.split(Xscaled_train, y_train)):

        #Fit the scaled training data to the ridge model
        ## START CODE ## (1 line of code)
        ridge.fit(Xscaled_train[train], y_train[train])
        ## END CODE ##

        #Calculate the average mean sqaured error
        ##  START CODE  ## (1 line of code)
        avg_mse += mean_squared_error(y_train[test], ridge.predict(Xscaled_train[test]))
        ##  END CODE    ##

    # Take the average mean squared error as metric
    ridge_avg_mse[alpha] = avg_mse / n_folds
```

```python
# Find the best value for alpha with minimum mean squared error
##  START CODE  ## (1 line of code)
best_alpha_ridge = min(ridge_avg_mse, key=ridge_avg_mse.get)
##  END CODE    ##

print(f"Best Alpha: {best_alpha_ridge}, Corresponding Average MSE: {ridge_avg_mse[best_alpha_ridge]}")
```

```python
X_train = X_train.to_numpy()
X_test = X_test.to_numpy()

lm_model = LinearRegression()
lm_model.fit(X_train, y_train)

y_pred_lm = lm_model.predict(X_test)

# Evaluate the model
mse_lm = mean_squared_error(y_test, y_pred_lm)
r2_lm = r2_score(y_test, y_pred_lm)

n_folds = 5
k_fold = KFold(n_splits=5, shuffle=True, random_state=50)
alphas = 10**np.linspace(-6, 6, 100)
lasso_avg_mse = {}
ridge_avg_mse = {}

# Iterate through alpha values for Lasso and Ridge
for alpha in alphas:
    lasso = Lasso(alpha=alpha)
    ridge = Ridge(alpha=alpha)
    mse_lasso = 0
    mse_ridge = 0
    for k, (train, test) in enumerate(k_fold.split(X_train, y_train)):
        lasso.fit(X_train[train], y_train[train])
        ridge.fit(X_train[train], y_train[train])
        mse_lasso  += mean_squared_error(y_train[test],lasso.predict(X_train[test]))
        mse_ridge  += mean_squared_error(y_train[test],ridge.predict(X_train[test]))
    lasso_avg_mse[alpha] = mse_lasso / n_folds
    ridge_avg_mse[alpha] = mse_ridge / n_folds

# Find the best alpha for Lasso and Ridge
best_alpha_lasso_NS = min(lasso_avg_mse, key=lasso_avg_mse.get)
best_alpha_ridge_NS = min(ridge_avg_mse, key=ridge_avg_mse.get)

print(f"Best Alpha for Lasso: {best_alpha_lasso_NS}, Corresponding Average MSE: {lasso_avg_mse[best_alpha_lasso_NS]}")
print(f"Best Alpha for Ridge: {best_alpha_ridge_NS}, Corresponding Average MSE: {ridge_avg_mse[best_alpha_ridge_NS]}")


# Lasso regression model with the best alpha
lasso_model = Lasso(alpha=best_alpha_lasso_NS)
lasso_model.fit(X_train, y_train)
y_pred_lasso = lasso_model.predict(X_test)
mse_lasso = mean_squared_error(y_test, y_pred_lasso)
r2_lasso = r2_score(y_test, y_pred_lasso)

# Ridge regression model with the best alpha
ridge_model = Ridge(alpha=best_alpha_ridge_NS)
ridge_model.fit(X_train, y_train)
y_pred_ridge = ridge_model.predict(X_test)
mse_ridge = mean_squared_error(y_test, y_pred_ridge)
r2_ridge = r2_score(y_test, y_pred_ridge)

r_squared_values = [r2_lm, r2_lasso, r2_ridge]
mse_values = [mse_lm, mse_lasso, mse_ridge]

# Create a DataFrame
final_report_not_scaled = pd.DataFrame({'Model': ['Linear', 'Lasso', 'Ridge'],
                   'R-squared': r_squared_values,
                   'MSE': mse_values})

display(final_report_not_scaled)
```

Why is it advisable to scale the independent variables when applying ridge or  lasso regression?

<font color='#4682B4'>

While applying lasso and ridge regression, it is always recommended to scale the variables because of the following reasons:

***1) Leveling the playing field***

Scaling ensures that all the features are on the same scale and hence regularization penalties are applied fairly based on true importance rather than arbitrary differences.

***2) Interpretability***

Scaled coefficients are more interpretable because their magnitudes directly indicate importance. Unscaled coefficients confound scale and importance.

***3) Convergence***

Scaling can improve the convergence properties of optimization algorithms used to fit Ridge and Lasso models. Variables on different scales can cause numerical instability during optimization, leading to slower convergence or non-convergence. Scaling helps stabilize the optimization process, making it more efficient.

In summary, scaling ensures the regularization penalties are applied meaningfully and coefficients are comparable and interpretable. This improves model stability, optimization, and understanding of feature effects in ridge and lasso regression.
